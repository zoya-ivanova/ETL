###3. Сделайте колонку purpose_category в которую войдут следующие категории:
• операции с автомобилем,<br>
• операции с недвижимостью,<br>
• проведение свадьбы,<br>
• получение образования

```python
df['purpose']
```

```python
data['purpose'].unique()
```

```python
# создаем функцию categorize_purpose()
def categorize_purpose(purpose):
  if 'авто' in purpose:
    return 'операции с автомобилем'
  if 'недвиж' in purpose or 'жиль' in purpose:
    return 'операции с недвижимостью'
  if 'свад' in purpose:
    return 'проведение свадьбы'
  if 'образов' in purpose:
    return 'получение образования'
```

```python
print(categorize_purpose('покупка жилья'))
```

```python
df['purpose_category'].value_counts()
```

```python
# примените функцию методом apply()
df['purpose_category'] = df['purpose'].apply(categorize_purpose)
df
```
